## __QUERY 1__

```python
c = con.cursor()
q = "SELECT c.customerName, p.productName, od.quantityOrdered, od.priceEach, (od.quantityOrdered*od.priceEach) AS totalPrice \
FROM orders o JOIN customers c \
ON c.customerNumber = o.customerNumber \
JOIN orderdetails od \
ON o.orderNumber = od.orderNumber \
JOIN products p \
ON od.productCode = p.productCode \
JOIN payments py \
ON py.customerNumber = c.customerNumber \
WHERE py.paymentDate = '2003-06-05' \
LIMIT 4;"
c.execute(q)
res = c.fetchall()
```

```python
df = pd.DataFrame(res, columns=c.column_names)
df
```

## __QUERY 2__

```python
c2 = con.cursor()
q2 = "SELECT o.orderNumber, c.customerName, c.city, c.country, (SUM(od.quantityOrdered)*SUM(od.priceEach)) AS totalPurchased FROM customers c JOIN orders o \
ON c.customerNumber = o.customerNumber \
JOIN orderdetails od \
ON o.orderNumber = od.orderNumber \
GROUP BY orderNumber \
ORDER BY totalPurchased DESC;"
c2.execute(q2)
res1 = c2.fetchall()
```

```python
df2 = pd.DataFrame(res1, columns=c2.column_names)
df2
```

## __QUERY 3__

```python
d = con.cursor()
qu = "SELECT p.productName, SUM(od.quantityOrdered) AS quantityOrdered FROM orderdetails od JOIN orders o \
ON od.orderNumber = o.orderNumber \
JOIN products p \
ON od.productCode = p.productCode \
WHERE o.orderDate BETWEEN '2003-01-05' AND '2003-12-10' \
GROUP BY productName \
ORDER BY quantityOrdered DESC \
LIMIT 10;"
d.execute(qu)
res2 = d.fetchall()
```

```python
df3 = pd.DataFrame(res2, columns=d.column_names)
df3
```
